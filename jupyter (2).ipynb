```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame
sns.set_style('whitegrid')  # Optional: Set the style for the plots

# Line chart for sleep schedule
def plot_sleep_schedule(df, time_column, sleep_column):
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df, x=time_column, y=sleep_column, marker="o", color="b", label="Sleep Duration")
    plt.title('Sleep Schedule Over Time', fontsize=14, weight='bold')
    plt.xlabel(time_column, fontsize=12)
    plt.ylabel('Sleep Duration (Hours)', fontsize=12)
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()  # Adjust layout for better visuals
    plt.show()

# Example usage (replace 'id' with a column representing time/order in your data)
plot_sleep_schedule(df, 'id', 'Sleep Duration')  # Replace 'id' with the appropriate time/order column
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to create a pie chart
def plot_pie_chart(df, column_name, title=None):
    # Get value counts for the column
    counts = df[column_name].value_counts()

    # Create the pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)
    plt.title(title if title else f'Distribution of {column_name}', fontsize=14, weight='bold')
    plt.show()

# Example usage for Financial Stress
plot_pie_chart(df, 'Financial Stress', title='Distribution of Financial Stress Levels')
```

```python
def plot_family_vs_finance_box(df):
    plt.figure(figsize=(8, 6))
    sns.boxplot(x="Family History of Mental Illness", y="Financial Stress", data=df, palette="coolwarm")
    plt.title("Distribution of Financial Stress by Family History of Mental Illness", fontsize=14, weight='bold')
    plt.xlabel("Family History of Mental Illness", fontsize=12)
    plt.ylabel("Financial Stress Level", fontsize=12)
    plt.show()

# Example usage
plot_family_vs_finance_box(df)
```

```python
def plot_family_vs_finance_pie(df):
    # Group the data by Family History and calculate the count of Financial Stress levels
    group_data = df.groupby(['Family History of Mental Illness', 'Financial Stress']).size().unstack()

    # Plot a stacked pie chart
    group_data.plot(kind='pie', subplots=True, figsize=(10, 8), autopct='%1.1f%%', startangle=140, colors=plt.cm.Set2.colors)
    plt.title("Financial Stress vs Family History of Mental Illness", fontsize=14, weight='bold')
    plt.show()

# Example usage
plot_family_vs_finance_pie(df)
```

```python
def plot_family_history_pie(df):
    counts = df['Family History of Mental Illness'].value_counts()
    plt.figure(figsize=(8, 8))
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)
    plt.title('Distribution of Family History of Mental Illness', fontsize=14, weight='bold')
    plt.show()

# Example usage
plot_family_history_pie(df)
```

```python
def plot_family_history_heatmap(df):
    heatmap_data = pd.crosstab(df['Family History of Mental Illness'], df['Depression'])
    plt.figure(figsize=(8, 6))
    sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='coolwarm', cbar=False)
    plt.title('Family History vs Depression Heatmap', fontsize=14, weight='bold')
    plt.xlabel('Depression', fontsize=12)
    plt.ylabel('Family History of Mental Illness', fontsize=12)
    plt.show()

# Example usage
plot_family_history_heatmap(df)
```

```python
def plot_work_study_histogram(df):
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df, x='Work/Study Hours', bins=10, kde=True, color='teal')
    plt.title('Distribution of Work/Study Hours', fontsize=14, weight='bold')
    plt.xlabel('Work/Study Hours', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.show()

# Example usage
plot_work_study_histogram(df)
```

```python
def plot_work_study_line_chart(df):
    avg_hours_by_age = df.groupby('Age')['Work/Study Hours'].mean()
    plt.figure(figsize=(8, 6))
    plt.plot(avg_hours_by_age.index, avg_hours_by_age.values, marker='o', color='purple')
    plt.title('Average Work/Study Hours by Age', fontsize=14, weight='bold')
    plt.xlabel('Age', fontsize=12)
    plt.ylabel('Average Work/Study Hours', fontsize=12)
    plt.grid(alpha=0.3)
    plt.show()

# Example usage
plot_work_study_line_chart(df)
```
